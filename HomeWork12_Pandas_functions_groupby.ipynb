### Задание 1

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг;
- оценка 4 и меньше - средний рейтинг;
- оценка 4.5 и 5 - высокий рейтинг.

Результат классификации запишите в столбец class.

```python
# импортируем программную библиотеку на языке Python для обработки и анализа данных
import pandas as pd 
```

```python
# загружаем датасет с оценками фильмов
ratings = pd.read_csv('ratings.csv')    
ratings.head()
```

```python
def classification_of_films(rating):
    
    """Классификация фильмов по правилам задания"""
    
    if rating <= 2.0:
        return 'low'
    elif 2.0 < rating <= 4.0:
        return 'avg'
    else:
        return 'high'
```

```python
# применяем метод apply к одному столбцу rating датасета ratings
ratings['class'] = ratings.rating.apply(classification_of_films)
ratings.head()
```

### Задание 2

Используем файл keywords.csv.Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'undefined'.Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

   'Центр': ['москва', 'тула', 'ярославль'],
   
   'Северо-Запад': ['петербург', 'псков', 'мурманск'],
   
   'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
   
}

Результат классификации запишите в отдельный столбец region.

```python
# загружаем датасет с поисковыми запросами
keywords = pd.read_csv('keywords.csv')
keywords.head()
```

```python
# гео-данные городов по регионам
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def geo_klassifikkator(keyword):
    
    """гео-классификатор запросов по региону"""
    
    for region, towns in geo_data.items(): # проходимся циклом по гео-данным с помощью метода items()
        for town in towns:                 # проходимся циклом по городам (значениям словаря)
            if town in keyword:            # если название города есть в столбце keyword датасета
                return region              # возвращаем название региона (ключ словаря)
    return 'undefined'                     # в противном случае, undefined - поисковый запрос не содержит названия города
```

```python
# применяем метод apply к одному столбцу keyword датасета keywords
keywords['region'] = keywords.keyword.apply(geo_klassifikkator)
keywords.head()
```

```python
# посмотрим, есть ли не'undefined' поисковые запросы
keywords[(keywords['region']!='undefined')]
```

### Задание 3

Есть мнение, что "раньше снимали настоящее кино, не то что сейчас". Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл ratings.csv из базы https://grouplens.org/datasets/movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

1. В переменную years запишите список из всех годов с 1950 по 2010.

2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

 - для каждой строки пройдите по всем годам списка years
 - если номер года присутствует в названии фильма, то функция возвращает этотгод как год выпуска
 - если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
 

3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец 'year'

4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца 'year' и отсортируйте результат по убыванию рейтинга

```python
# загружаем датасет
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
# с помощью функции range() заапишем в переменную years список list() из всех годов с 1950 по 2010
# генерим числа в заданном диапозоне и с помощью функции list() создадим список
years = list(range(1950,2011))
```

```python
def production_year(row):
    
    """функция, которая каждой строке из названия фильма выставляет год выпуска"""
    
    for year in years:         # проходимся циклом по каждому элементу списка years
        if str(year) in row:   # если номер года присутствует в названии фильма,    
            return year        # функция возвращает этот год как год выпуска
    return 1900                # в противном случае, возвращается 1900 год
```

```python
# применяем метод apply к одному столбцу title датасета movies
movies['year']= movies.title.apply(production_year)
movies
```

```python
# посчитаем средний рейтинг всех фильмов, для этого с помощью функции merge() соединяем датасеты movies и ratings по ключу movieId
rating_film = ratings.merge(movies, on='movieId', how='left')
rating_film
```

```python
# делаем группировку с помощью функции groupby() по столбцу year
# считаем среднее значение с помощью функции mean() по столбцу rating
# сортируем результат по убыванию рейтинга с помощью функции sort_values
rating_film.groupby('year').mean().rating.sort_values(ascending=False)
```
