## Задание 2

Дана переменная, в которой хранится список поисковых запросов пользователя. Вам необходимо написать программу, которая выведет на экран распределение количества слов в запросах в требуемом виде.

```python
# 1 вариант решения (плохой вариант, т.к. количество слов в запросе проставлять в ручную; сработает на маленьком списке)

queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]
res = []
for el in queries:                   # Перебираем элементы списка - строки поисковых запросов
    if el not in res:
        res.append(len(el.split()))  # Добавляем в новый список только длину строки, разбитую пробелами по умолчанию
# print(res)
# print(f'В списке всего {len(res)} поисковых запросов')
print(f'Поисковых запросов, содержащих 2 слов(а): {round(res.count(2) / len(res) * 100, 2)}%')
print(f'Поисковых запросов, содержащих 3 слов(а): {round(res.count(3) / len(res) * 100, 2)}%')
```

```python
# 2 вариант решения (хороший вариант, программа считает все сама)

queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]

# queries = [
#     'смотреть сериалы онлайн',
#     'новости спорта',
#     'афиша кино',
#     'курс доллара',
#     'сериалы этим летом',
#     'курс по питону',
#     'сериалы про спорт',
#     'смотреть кино в кинотеатре',
#     'участвовать в марафоне по питону',
#     'отдохнуть'
# ]

res = {}

for el in queries:             # пройдем циклом по элементам списка queries (элементы - строки)
    len_el = len(el.split())   # задаем переменную, опеределяющую длину элемнтов (длину строковых запросов, разбитых пробелами по умолчанию)
#     print(len_el)
    if len_el not in res:      # если переменной нет в нашем словаре, то добавляем (ключ - переменная, значение - 1, встречается первый раз)
        res[len_el] = 1
    else:                      # если такой ключ в словаре уже есть, значение ключа увеличивается на 1
        res[len_el] += 1
# res

for k, v in res.items():       # пройдемся циклом for по ключам и значениям (при помощи метода .items()) словаря res
                               # для каждой пары ключ-значение выведем распределение количества слов в запросах в требуемом виде
    
    print(f'Поисковых запросов, содержащих {k} слов(а): {round(v/len(queries)*100,2)} %')
    
```

## Задание 3

Дана переменная, в которой хранится информация о затратах и доходе рекламных кампаний по различным источникам. Необходимо дополнить исходную структуру показателем ROI, который рассчитаем по формуле: (revenue / cost - 1) * 100.

```python
results = {
    'vk': {'revenue': 103, 'cost': 98},
    'yandex': {'revenue': 179, 'cost': 153},
    'facebook': {'revenue': 103, 'cost': 110},
    'adwords': {'revenue': 35, 'cost': 34},
    'twitter': {'revenue': 11, 'cost': 24},
}

for el in results.values():                                      # перебираем значения (вложенные словари) исходного словаря
    el['ROI'] = round((el['revenue'] / el['cost'] - 1) * 100, 2) # во вложенные словари добавляем элемент (ключ ROI, значение высчитываем по формуле)
# results                                                        # выводим словарь results с дополненным в исходную структуру показателем ROI

{k: v for k, v in sorted(results.items())}                       # создадим новый словарь из словаря results, отсортировав (при помощи функции sorted) список пар ключ-значение(методом .items())
```

## Задание 4

Дана переменная, в которой хранится статистика рекламных каналов по объемам продаж (пример структуры данных приведен ниже). Напишите программу, которая возвращает название канала с максимальным объемом продаж.

```python
stats = {'facebook': 55, 'yandex': 115, 'vk': 120, 'google': 99, 'email': 42, 'ok': 98}

for k, v in stats.items():       # пройдемся циклом for по ключам и значениям (при помощи метода .items()) словаря stats
    if v == max(stats.values()): # для максимального значения (при помощи метода .values()) выводим его ключ
        
        print(f'Максимальный объем продаж на рекламном канале: {k}')
```

## Задание 5

Дан список произвольной длины. Необходимо написать код, который на основе исходного списка составит словарь такого уровня вложенности, какова длина исхондого списка.

```python
# my_list = ['2018-01-01', 'yandex', 'cpc', 100]
my_list = ['a', 'b', 'c', 'd', 'e', 'f']

res = my_list[-1]                    # задаем новый список с одним элементом - последний элемент исходного списка

for el in (my_list[:-1])[::-1]:      # проходимся циклом for по исходному списку my_list в обратном порядке (используя синтаксис среза [::-1]), удаляя последний элемент при помощи среза [:-1]
# for el in reversed(my_list[:-1]):  # или используя функцию reversed()
    res = {el:res}                   # получаем словари, ключ - элемент заданного перевернутого спсика, а значение - последенее значение списка res

print(res)
```

## Задание 6

Дана книга рецептов с информацией о том, сколько ингредиентов нужно для приготовления блюда в расчете на одну порцию (пример данных представлен ниже).
Напишите программу, которая будет запрашивать у пользователя количество порций для приготовления этих блюд и отображать информацию о суммарном количестве требуемых ингредиентов в указанном виде.

```python
cook_book = {
  'салат': [
     {'ingridient_name': 'сыр', 'quantity': 50, 'measure': 'гр'},
     {'ingridient_name': 'томаты', 'quantity': 2, 'measure': 'шт'},
     {'ingridient_name': 'огурцы', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'маслины', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'оливковое масло', 'quantity': 20, 'measure': 'мл'},
     {'ingridient_name': 'салат', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'перец', 'quantity': 20, 'measure': 'гр'}
    ],
  'пицца': [
     {'ingridient_name': 'сыр', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'колбаса', 'quantity': 30, 'measure': 'гр'},
     {'ingridient_name': 'бекон', 'quantity': 30, 'measure': 'гр'},
     {'ingridient_name': 'оливки', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'томаты', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'тесто', 'quantity': 100, 'measure': 'гр'},   
    ],
  'лимонад': [
     {'ingridient_name': 'лимон', 'quantity': 1, 'measure': 'шт'},
     {'ingridient_name': 'вода', 'quantity': 200, 'measure': 'мл'},
     {'ingridient_name': 'сахар', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'лайм', 'quantity': 20, 'measure': 'гр'},    
    ]
}

n = int(input('Введите количество порций: '))
res = {}                            # создаем пустой словарь res

for el in cook_book.values():       # перебираем значения (вложенные словари - ингридиенты блюд) исходного словаря
    for m in el:                    # пройдемся циклом по вложенным словарям
        a = m['ingridient_name']    # присваиваем значения переменным
        c = m['measure']
        b = m['quantity']*n         # умножаем количество продуктов на количество персон
        if (a,c) not in res:        # заполняем словарь: ключ - кортеж (ингредиент, мера), значение - количество
            res[a,c] = b            # если блюдо вида (конкретный ингридиент, мера веса) нет в словаре, добавляем ключ-значение
        else:                       # в противном случае, при повторяющемся ключе - складываем количество
            res[a,c] += b
for (a,c),b in sorted(res.items()): # из получившегося словаря res(при помощи функции sorted и метода .items() сортируем ключи) выводим список ингредиентов в заданном виде
    print(f'{a}: {b} {c}')
```
