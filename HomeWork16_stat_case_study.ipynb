### Задание 2. Работа с выбросами

В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
# Рассмотрим выбросы по величине rectal_temperature

q1 = df_horse.rectal_temperature.quantile(0.25)
q3 = df_horse.rectal_temperature.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)

outliers = df_horse[(df_horse.rectal_temperature > upper_bound) | (df_horse.rectal_temperature < lower_bound)].sort_values('rectal_temperature')
print(outliers)
```

```python
# Рассмотрим выбросы по величине pulse

q1 = df_horse.pulse.quantile(0.25)
q3 = df_horse.pulse.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)

outliers = df_horse[df_horse.pulse.between(lower_bound, upper_bound, inclusive=True)]
print(outliers)
```

Выбросы присутствуют в данных rectal_temperature, pulse, но могут быть признаны естественными для данной выборки.
