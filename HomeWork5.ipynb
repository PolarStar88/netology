# Домашнее задание к лекции "Функции"

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Общие требования к программе:

- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением documents и directories);
- пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды "q").

```python
# перечень всех документов
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

```python
# перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

## Задание 1

```python
def owner_doc():
    
    """
    Пользователь по команде "p" может узнать владельца документа по его номеру
    """
    
    input_number = input('Введите номер документа: ')
    name = None                                       # случай отсутствующего имени
    for el in documents:                              # пройдемся циклом по перечню всех документов
        if input_number == el['number']:              # в случае, если введенный пользователем номер документа внутри перечня
            name = el['name']                         # и имя владельца существует
            print(f'Владелец документа: {name}')      # выводим имя владельца
    if input_number != el['number'] and name is None: # в противном случае, документ не существует
        print('Документ не существует.')
        
owner_doc()
```

```python
def shelf_doc():
    
    """
    Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
    """
    
    input_number = input('Введите номер документа: ')
    doc_found = 0                                          # вводим счетчик поиска, если номер документа есть на полке
    for k, v in directories.items():                       # проходимся циклом по перечню полок (с помощью метода .items() опеределяем ключ - номер полки, значение - номер документа)
        if input_number in v:                              # если введенный пользователем номер документа в перечне полок,
            doc_found += 1                                 # накручиваем счетчик на 1
            number_shelf = k                               # номер полки - ключ
    if doc_found == 1:                                     # как только документ находится на полке, выводим номер полки
        print('Документ хранится на полке:', number_shelf)
    else:                                                  # в противном случае, документа нет
        print('Документ не найден в базе')

shelf_doc()
```

```python
def info_doc():
    
    """
    Пользователь по команде "l" может увидеть полную информацию по всем документам
    """
    
    for el in documents:                 # проходимся циклом по перечню всех документов
        for k, v in directories.items(): # проходимся циклом по перечню полок
            if el['number'] in v:        # если номер документа есть и в перчне документов, и в перечне полок,
                number_shelf = k         # номер полки - ключ
                                         # и выводим полную информацию по всем имеющимся документам в перечне
        print('№: ', el['number'], ';', 'тип: ', el['type'], ';', 'владелец: ', el['name'], ';' , 'полка хранения: ', number_shelf)
        
info_doc()
```

```python
def add_new_shelf():
    
    """
    Пользователь по команде "ads" может добавить новую полку
    """
    
    input_shelf = input('Введите номер новой полки: ')
    if input_shelf in directories.keys():                                                   # если введенный пользователем номер полки есть в перечне полок,
        print('Полка с таким номером уже существует!', 'Текущий перечень полок: ', end='')  
        print(*list(directories.keys()), sep=', ')                                          # выводим текущий список полок, распаковывая итерируемые объекты
    else:
        directories[input_shelf] = []                                                       # в противном случае, добавляем номер полки в текущий список
        print(f'Полка номер {input_shelf} добавлена.', 'Текущий перечень полок: ', end='')  # и выводим список
        print(*list(directories.keys()), sep=', ')

add_new_shelf()
```

```python
def delete_empty_shelf():
    
    """
    Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
    """
    
    shelf_delete = input('Введите номер полки: ')
    if shelf_delete in list(directories.keys()) and directories[shelf_delete] == []:                         # если номер полки, которую нужно удалить, в списке перечня полок и полка пуста
        del directories[shelf_delete]                                                                        # удаляем полку с помощью метода del
        print('Полка удалена. Текущий перечень полок: ', end='')
        print(*list(directories.keys()), sep=', ')                                                           # выводим текущий список полок, распаковывая итерируемые объекты
    elif shelf_delete in directories and directories[shelf_delete] != []:                                    # если полка в списке перечня полоко и она не пуста
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', end='') # выводим предостережение, нужно удалить документы с полки
        print(*list(directories.keys()), sep=', ')
    else:                                                                                                     # в противном случае, введенного номера полки не существует
        print('Такой полки не существует. Текущий перечень полок: ', end='')
        print(*list(directories.keys()), sep=', ')
        
delete_empty_shelf()
```

```python
def main():                                       # для удобства пользовательского ввода, объединим все функции в одну с помощью цикла while
    while True:
        user_input = input('Введите команду ')
        if user_input == 'p':
            owner_doc()
        elif user_input == 's':
            shelf_doc()
        elif user_input == 'l':
            info_doc()
        elif user_input == 'ads':
            add_new_shelf()
        elif user_input == 'ds':
            delete_empty_shelf()
        elif user_input == 'q':
            print('Работа завершена.')
            break                                  # завершим работу цикла после введения команды q
        else:
            print('Введена неверная команда')
            
main()
```
